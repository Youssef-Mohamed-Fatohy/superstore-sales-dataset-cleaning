```python
import pandas as pd

# 1. Load the dataset
df = pd.read_csv('Superstore Sales Dataset.csv')
```

```python
# 2. Handle Missing Values
df.loc[(df['City'] == 'Burlington') & (df['State'] == 'Vermont') & (df['Postal Code'].isnull()), 'Postal Code'] = 5401.0
```

```python
# 3. Convert Date Columns to Datetime Objects
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%d/%m/%Y')
```

```python
# 4. Standardize Postal Code
df['Postal Code'] = df['Postal Code'].astype(int).astype(str).str.zfill(5)
```

```python
# 5. Text Cleaning
str_cols = df.select_dtypes(include=['object']).columns
for col in str_cols:
    df[col] = df[col].str.strip()
```

```python
# 6. Normalize Column Names
# Replace spaces with underscores and convert to lowercase for easier coding.
df.columns = [col.replace(' ', '_').lower() for col in df.columns]
```

```python
# 7. Final Verification
print("Missing values after cleaning:\n", df.isnull().sum())
print("\nData types after cleaning:\n", df.dtypes)
```
